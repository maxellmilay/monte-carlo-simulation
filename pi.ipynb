## Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Initialize Variables

```python
# initialize variables
n_simulations = 100000
n_points_circle = 0
n_points_square = 0

# points in the circle
circle_x = []
circle_y = []

# points outside the circle
outside_x = []
outside_y = []
```

## Run Simulation

```python
for i in range(n_simulations):
    
    # x is randomly drawn from a continuous uniform distritbuion
    x = np.random.uniform(-1, 1)
    
    # y is randomly drawn from a continuous uniform distribution
    y = np.random.uniform(-1, 1)
    
    # calculate the distance between the point and the origin
    dist_from_origin = x**2 + y**2
    
    # if the distance is smaller than or equal to 1, the point is in the circle
    if dist_from_origin <= 1:
        circle_x.append(x)
        circle_y.append(y)
        n_points_circle += 1
    else:
        # print(x,y,dist_from_origin)
        outside_x.append(x)
        outside_y.append(y)
    
    # by definition of the uniform distribution, the point is in the square
    n_points_square += 1
```

```python
print("Inside circle: ",len(circle_x), len(circle_y))
print("Outside circle: ",len(outside_x), len(outside_y))
```

## Visualize Simulation

```python
plt.scatter(circle_x, circle_y, label='Inside Circle', s=1, color='Blue')
plt.scatter(outside_x, outside_y, label='Outside Circle', s=1, color='Red')
plt.title('Simulated Random Values')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
